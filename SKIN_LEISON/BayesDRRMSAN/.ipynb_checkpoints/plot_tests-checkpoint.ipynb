```python
import json
import matplotlib.pyplot as plt

with open('history_skin_drrmsan.json', 'r') as f:
    array = json.load(f)
print (array)
```

```python
accuracy = []
for item in array['accuracy']:
  accuracy.append(array['accuracy'][item])
val_accuracy = []
for item in array['val_accuracy']:
  val_accuracy.append(array['val_accuracy'][item])
#print(accuracy)
```

```python
plt.title('accuracy')
plt.plot(accuracy,color='green')
plt.plot(val_accuracy,color='yellow')
```
