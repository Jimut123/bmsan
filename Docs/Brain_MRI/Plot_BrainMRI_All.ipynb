```python
import json
import matplotlib.pyplot as plt
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
with open('history_attnR2Unet_brainMRI.json', 'r') as f:
    attnR2Unet = json.load(f)
print (len(attnR2Unet))
```

```python

with open('history_r2unet_brainmri.json', 'r') as f:
    r2unet = json.load(f)
print (len(r2unet))

```

```python

with open('history_attn_unet_brainMRI.json', 'r') as f:
    attn_unet = json.load(f)
print (len(attn_unet))
```

```python

with open('history_multi_res_unet_brainmri.json', 'r') as f:
    multi_res_unet = json.load(f)
print (len(multi_res_unet))
```

```python

with open('history_UNet_brainmri.json', 'r') as f:
    UNet = json.load(f)
print (len(UNet))
```

```python

with open('history_orunet_brainmri.json', 'r') as f:
    orunet = json.load(f)
print (len(orunet))
```

```python
"""
attnR2Unet
r2unet
attn_unet
multi_res_unet
UNet
orunet
"""
plt.rc('font',family='Times New Roman')

accuracy_attnR2Unet = []
for item in attnR2Unet['accuracy']:
    accuracy_attnR2Unet.append(attnR2Unet['accuracy'][item])

accuracy_r2unet = []
for item in r2unet['accuracy']:
    accuracy_r2unet.append(r2unet['accuracy'][item])

accuracy_attn_unet = []
for item in attn_unet['accuracy']:
    accuracy_attn_unet.append(attn_unet['accuracy'][item])

accuracy_multi_res_unet = []
for item in multi_res_unet['accuracy']:
    accuracy_multi_res_unet.append(multi_res_unet['accuracy'][item])

accuracy_UNet = []
for item in UNet['accuracy']:
    accuracy_UNet.append(UNet['accuracy'][item])
    
accuracy_orunet = []
for item in orunet['accuracy']:
    accuracy_orunet.append(orunet['accuracy'][item])

plt.figure(figsize=(10,6))
plt.title('Accuracy for Brain MRI dataset',fontsize=20, fontname = 'Times New Roman', fontweight = 500)
#plt.plot(accuracy_attnR2Unet,color='green')
#plt.plot(accuracy_r2unet,color='blue')
#plt.plot(accuracy_attn_unet,color='orange')
plt.plot(accuracy_multi_res_unet,color='yellow')
plt.plot(accuracy_UNet,color='black')
#plt.plot(accuracy_orunet,color='purple')

plt.legend(['AttnR2Unet','R2Unet','AttnUnet','MultiResUnet','Mod UNet','UNet'],loc="lower right", 
           prop={'family':'Times New Roman', 'size':20})
plt.savefig('brain_mri_accuracy_coef.eps')
```

```python
"""
attnR2Unet
r2unet
attn_unet
multi_res_unet
UNet
orunet
"""
print("Accuracy for attnR2Unet = ",accuracy_attnR2Unet[-1])
print("Accuracy for r2unet = ",accuracy_r2unet[-1])
print("Accuracy for attn_unet = ",accuracy_attn_unet[-1])
print("Accuracy for multi_res_unet = ",accuracy_multi_res_unet[-1])
print("Accuracy for UNet = ",accuracy_UNet[-1])
print("Accuracy for orunet = ",accuracy_orunet[-1])
```
