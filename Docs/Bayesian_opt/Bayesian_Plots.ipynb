```python
import matplotlib.pyplot as plt
```

```python
file1=open("bayesian_opt_skin_lesion.txt","r")

alphas = []
alpha_1 = []
dices = []
counter = []
count = 1

get_dict = {}
# Iterate over each line in the file
dummy = 0
for line in file1.readlines():
    dummy += 1
    if dummy <= 2:
        continue
    # Separate each item in the line
    items=line.split()
    alpha = items[:-1]
    alpha_1.append(items[0])
    #print(alpha)
    alphas.append(alpha)
    #print(items)
    dice = items[-1]
    get_dict[float(dice)] = alpha
    counter.append(count)
    count+=1
    #print(dice)
    dices.append(float(dice))
    
file1.close()
```

```python
# max dice with corresponding values of alphas
# alpha_1 = least scale, alpha_4 = same scale as I
max(get_dict.items(), key=lambda get_dict: get_dict[1])
```

```python
#fig = plt.figure()
# ax = plt.subplot(111)
plt.figure(figsize=(15,6))
plt.title('Dice Vs Alphas for Skin Lesion dataset', fontsize=20, fontname = 'DejaVu Serif', fontweight = 500)
plt.plot(counter, dices, linewidth=3.0, color='black')
plt.plot(counter, list(float(item[0]) for item in alphas), linewidth=2.0, linestyle='--')
plt.plot(counter, list(float(item[1]) for item in alphas), linewidth=2.0, linestyle='--',)
plt.plot(counter, list(float(item[2]) for item in alphas), linewidth=2.0, linestyle='--',)
plt.plot(counter, list(float(item[3]) for item in alphas), linewidth=2.0, linestyle='--',)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.xlabel("Iterations", fontsize=20, fontname = 'DejaVu Serif', fontweight = 500)
plt.ylabel("Values", fontsize=20, fontname = 'DejaVu Serif', fontweight = 500)
lgd = plt.legend(['dice', r'$\alpha_1$', r'$\alpha_2$', r'$\alpha_3$', r'$\alpha_4$'],loc="upper left",
          prop={'family':'Times New Roman', 'size':20}, bbox_to_anchor=(1, 1))

#plt.show()
plt.savefig('Bayesian_skin_lesion.eps',  bbox_extra_artists=(lgd,), bbox_inches='tight')
```
