```python
import matplotlib.pyplot as plt
```

```python


file1=open("bayesian_opt_skin_lesion.txt","r")

alphas = []
alpha_1 = []
dices = []
counter = []
count = 1
# Iterate over each line in the file
for line in file1.readlines():
    # Separate each item in the line
    items=line.split()
    alpha = items[:-1]
    alpha_1.append(items[0])
    #print(alpha)
    alphas.append(alpha)
    #print(items)
    dice = items[-1]
    counter.append(count)
    count+=1
    #print(dice)
    dices.append(float(dice))
    
file1.close()
```
